```python
import pandas as pd
import requests
import re
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
from datetime import datetime
```

```python
url = 'https://www.investing.com/commodities/brent-oil-historical-data'
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_5)'
           'AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.102 Safari/537.36'}
html = requests.get(url, headers=headers)
soup = BeautifulSoup(html.content,'html.parser')
```

```python
html
```

```python
title=soup.find_all('h2',{'class':'float_lang_base_1 inlineblock'})
```

```python
title=title[0].get_text()
title
```

```python
tables = soup.find('table',id="curr_table")
arrays = []
number = 1
array1 = []
array2 = []
tables
```

```python
for table in tables.find_all('tr'):
    
    headers=table.find_all('th')
    for header in headers:
        headerspacing = header.get_text()
        array2.append(headerspacing)
        if number == 7:
            arrays.append(array2)
            array2 = []
            number = 0
        number+=1
        
    columns = table.find_all('td')
    for column in columns:
        columnspacing = column.get_text()
        array1.append(columnspacing)
        if number == 7:
            arrays.append(array1)
            array1 = []
            number = 0
        number+=1
```

```python
arrays = np.array(arrays)
```

```python
df = pd.DataFrame(arrays)
new_header = df.iloc[0] 

df = df[1:] 

df.columns = new_header
df
```

```python
now = datetime.now()
now = now.strftime("%Y-%m-%d-%H-%M-%S")
filename = "brent-oil-historical-data-"+str(now)+".csv"
df.to_csv(filename,index=False)
```
