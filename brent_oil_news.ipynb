```python
import pandas as pd
import requests
import re
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
from datetime import datetime
```

```python
arrays = []
for number in range(1, 5):
    url = 'https://www.investing.com/commodities/brent-oil-news/'+str(number)
    #print('url', url)
    headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_5)'
               'AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.102 Safari/537.36'}
    html = requests.get(url, headers=headers)
    soup = BeautifulSoup(html.content,'html.parser')
    
    articles = soup.find_all('div',{'class':'mediumTitle1'})
    articles = articles[1]
    
    for article in articles.find_all('article', {'class':'articleItem'}):
        article_csv = []
        textDiv=article.find('div',{'class':'textDiv'})
        header=textDiv.find('a',{'class':'title'})
        headerspacing = header.get_text()
        link = header.get('href')

        #articleDetail=article.find('span',{'class':'articleDetails'})
        articleDate=article.find('span',{'class':'date'})
        date = articleDate.get_text()
        #print('articleDetail', headerspacing, link, date)

        article_csv.append(headerspacing)
        article_csv.append(link)
        article_csv.append(date)

        arrays.append(article_csv)
        
arrays = np.array(arrays)       
arrays
    
df = pd.DataFrame(arrays)
df.columns = ["Title", "Link", "Created Date"]
df
```

```python
now = datetime.now()
now = now.strftime("%Y-%m-%d-%H-%M-%S")
filename = "brent-oil-news-"+str(now)+".csv"
df.to_csv(filename,index=False)
```
